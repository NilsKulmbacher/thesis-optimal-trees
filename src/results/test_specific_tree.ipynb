## Validation class for specific tree

For manual validation of generated trees and metrics, a specific tree can be codified such that metrics (e.g accuracy) can be tested.

```python
import sys
sys.path.append("..")
import preprocessing.Datasets as DS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
problem = DS.skinnonskin 
pname = problem.__name__.upper()
print(f'---{pname}---')


df = problem('../../datasets/original/')
X_train, X_test, y_train, y_test = train_test_split(df.drop('y', axis=1), df['y'], test_size=0.2, random_state=42)
```

```python
# Manual prediction of sample GOSDT decision tree
def predict_decision_tree(df):
    predictions = []
    
    for _, row in df.iterrows():
        if row['X_0'] <= 176.5:
            if row['X_2'] <= 137.5:
                predictions.append(1)
            else:
                predictions.append(0)
        else:
            predictions.append(1)
    
    return predictions

data = X_train[['X_0', 'X_2']]
prediction = predict_decision_tree(data)
print(prediction)
```

```python
# Calculate accuracy for validation
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, prediction))
```
